## 5.1 Organização e descrição da telemetria

Protocolo de Telemetria: Critérios de Decolagem (DECOLAR / ABORTAR)

Este protocolo define os limites operacionais estritos para a autorização de lançamento da nave **Aurora Siger**. Qualquer valor fora das faixas especificadas abaixo resulta em **Abortar decolagem**.

| Variável | Descrição | Parâmetro Nominal (DECOLAR) | Ação em Caso de Falha (ABORTAR) |
| :--- | :--- | :--- | :--- |
| `temp_interna` | Temperatura do módulo de comando. | **18°C a 26°C** | ABORTAR: Risco à integridade do hardware/tripulação. |
| `temp_externa` | Temperatura da fuselagem externa. | **-100°C a 150°C** | ABORTAR: Condições ambientais fora do limite de proteção. |
| `integridade` | Sensores de integridade estrutural. | **1 (True)** | ABORTAR: Falha estrutural ou fissura detectada. |
| `energia` | Nível de carga das baterias principais. | **> 80%** | ABORTAR: Autonomia insuficiente para a subida. |
| `pressao_psi` | Pressão interna dos tanques (PSI). | **450 a 550 PSI (Pound-Force per Square Inch ou Libra-Força por Polegada Quadrada)** | ABORTAR: Risco de explosão ou falha de ignição. |
| `status` | Diagnóstico de módulos críticos. | **1 (True)** | ABORTAR: Falha em subsistema vital ou perda de sinal. |

A organização da telemetria foi estruturada com o objetivo de representar, os principais parâmetros para a validação da decolagem da nave.

A variável temp_interna representa a temperatura do módulo de comando, sendo um indicador das condições internas da nave. Foi definido um intervalo entre 18°C e 26°C, garantindo segurança tanto para os sistemas eletrônicos quanto para a tripulação. Valores fora dessa faixa indicam risco de superaquecimento ou falha operacional, resultando na interrupção da missão.

Já a temp_externa reflete as condições ambientais enfrentadas pela fuselagem. Como o ambiente externo é mais extremo, foi estabelecida uma faixa mais ampla, entre -100°C e 150°C. Valores fora desse intervalo podem comprometer a estrutura da nave ou a eficiência dos sensores.

A variável integridade é um indicador binário que representa a condição estrutural da nave. O valor 1 indica que não há falhas detectadas, enquanto o valor 0 representa problemas estruturais. Por se tratar de um fator crítico, qualquer inconsistência resulta automaticamente na interrupção da missão.

O nível de energia corresponde à carga disponível nas baterias principais. Foi definido um valor mínimo superior a 80%, assegurando autonomia suficiente para suportar as demandas energéticas da decolagem. Níveis abaixo desse limite indicam risco operacional.

A variável pressao refere-se à pressão interna dos tanques de combustível, sendo um parâmetro fundamental para o funcionamento adequado dos motores. A faixa segura foi definida entre 450 e 550 PSI, pois valores fora desse intervalo podem causar falhas de ignição ou até riscos de explosão.

Por fim, o status dos módulos críticos também é representado de forma binária, indicando se os subsistemas essenciais estão operacionais. Um valor verdadeiro (1) garante o funcionamento adequado, enquanto qualquer falha implica na impossibilidade de prosseguir com a decolagem.

## 5.2 Algoritmo de Verificação

### Pseudocódigo

FUNÇÃO analisar_componentes(df):

## 5.3 Script em Python

```python
import pandas as pd
```

```python
# Função para análise de componentes da nave e seus status final (ABORTAR/DECOLAR)
# Essa função vai levar como base o protócolo decidido para o lançamento ou não da nave Aurora. Fazendo todas as verificações necessárias 
# o método cria duas colunas "Obs" e "decolagem". A coluna "obs" anotará todas as falhas notadas e o campo de decolagem trará o status final.
def analisar_componentes(df: pd.DataFrame):
    df['Obs'] = ""
    df['decolagem'] = "DECOLAR"

    for i in range(len(df)):
        obs = []
        abortar = False
        # Temp_interna deve ser >= 18°C e <= 26°C:
        if df.loc[i, "temp_interna"] < 18 or df.loc[i, "temp_interna"] > 26:
            abortar = True
            obs.append("Risco à integridade do hardware/tripulação")
        # Temp_externa deve ser >= -100°C e <= 150°C
        if df.loc[i, 'temp_externa'] < -100 or df.loc[i, 'temp_externa'] > 150:
            abortar = True
            obs.append("Condições ambientais fora do limite de proteção")
        # Integridade deve ser = 1
        if df.loc[i, "integridade"] != 1:
            abortar = True
            obs.append("Falha estrutural ou fissura detectada")
        # Energia deve ser > 0.80 (80%)
        if df.loc[i, 'energia'] <= 0.80:
            abortar = True
            obs.append("Autonomia insuficiente para a subida")
        # Pressão deve ser >= 450 e <= 550
        if df.loc[i, 'pressao_psi'] < 450 or df.loc[i, 'pressao_psi'] > 550:
            abortar = True
            obs.append("Risco de explosão ou falha de ignição")
        # Status deve ser = 1
        if df.loc[i, 'status'] != 1:
            abortar = True
            obs.append("Falha em subsistema vital ou perda de sinal")

        if abortar:
            df.loc[i, 'decolagem'] = "ABORTAR"
            df.loc[i, 'Obs'] = ", ".join(obs)
        else:
            df.loc[i, 'Obs'] = "OK"

    return df


# Leitura da base de dados
df_nave = pd.read_csv('dados_nave.csv',sep=',')
# Atualizando a base
df_nave_analizado = analisar_componentes(df_nave)
display(df_nave_analizado)
```

## 5.4 Análise Energética

Capacidade Total = 190 kwh;
Consumo estimado = 20 kwh;
Perdas energéticas = 10% 

ENERGIA > 120 = "SUFICIENTE";
ENERGIA > 60 AND ENERGIA < 120 = "ALERTA";
ENERGIA < 60 = "INSUFICIENTE"


A análise energética foi estruturada com o objetivo de avaliar se a nave possui autonomia suficiente para realizar a decolagem com segurança, considerando o consumo operacional e as perdas.

Inicialmente, foi definida a capacidade total da nave em 190 kWh, representando o limite máximo de armazenamento energético disponível. A partir disso, considerou-se um consumo estimado de 20 kWh durante o processo de decolagem.

Além disso, foi incorporado ao modelo uma perda energética de 10%, refletindo ineficiências naturais do sistema, como dissipação térmica, conversões de energia, etc.

Com base nesses parâmetros, foram definidos três níveis de classificação energética. Valores de energia superiores a 120 kWh foram considerados “suficientes”, garantindo uma margem de segurança adequada mesmo após o consumo e as perdas previstas. Já valores abaixo de 120 kWh foram classificados como “alerta”, indicando que, embora a nave ainda possa operar, há redução na margem de segurança energética. Por fim, níveis inferiores a 60 kWh foram considerados “insuficientes”, representando um cenário crítico em que a energia disponível não é capaz de sustentar a operação de forma segura.

Apesar dessa análise mais detalhada sobre a energia das baterias, foi definido anteriormente que apenas níveis acima de 80% poderiam resultar na decolagem da nave, sendo assim, baterias em níveis de "ALERTA" e "INSUFICIENTE" não vão decolar.

```python
capacidade_total = 190
perda_energetica = 0.1
consumo_estimado = 20

for i in range(0,len(df_nave_analizado)):
    # Calcular energia disponível
    energia_disp = capacidade_total * df_nave_analizado.loc[i,"energia"]
    
    # Perdas energéticas
    energia_disp = energia_disp * (1 - perda_energetica)
    
    # Consumo da embalagem
    energia_disp = energia_disp - consumo_estimado
    
    df_nave_analizado.loc[i,"energia_restante_kwh"]= round(energia_disp,2)
    
    # Status da energia restante
    # Maior que 120 = SUFICIENTE
    if df_nave_analizado.loc[i,'energia_restante_kwh'] > 120:
        df_nave_analizado.loc[i,'status_energia'] = 'SUFICIENTE'
    # Maior que 60 = ALERTA
    elif df_nave_analizado.loc[i,'energia_restante_kwh'] > 60:
        df_nave_analizado.loc[i,'status_energia']='ALERTA'
    # Menor que 60 = INSUFICIENTE
    else:
        df_nave_analizado.loc[i,'status_energia']='INSUFICIENTE'
        
    
display(df_nave_analizado)
```
